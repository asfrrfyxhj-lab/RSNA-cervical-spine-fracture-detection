```python
import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import numpy as np
import cv2
import timm  
from tqdm import tqdm
# [Ï∂îÍ∞Ä] f1_scoreÎ•º ÌèâÍ∞Ä ÏßÄÌëú ÎùºÏù¥Î∏åÎü¨Î¶¨ÏóêÏÑú Ï∂îÍ∞ÄÎ°ú Î∂àÎü¨ÏòµÎãàÎã§.
from sklearn.metrics import roc_auc_score, accuracy_score, recall_score, precision_score, f1_score
import albumentations as A
from albumentations.pytorch import ToTensorV2
import matplotlib.pyplot as plt 

# --- [1. ÏãúÏä§ÌÖú ÏóêÎü¨ Î∞è Î©àÏ∂§ Î∞©ÏßÄ] ---
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
cv2.setNumThreads(0) 

# ==========================================
# 2. ÏÑ§Ï†ï (Configuration)
# ==========================================
CONFIG = {
    'base_dir': r'C:\conda\3. Project\rsna-2022-cervical-spine-fracture-detection\golden_dataset',
    'manifest_path': r'C:\conda\3. Project\rsna-2022-cervical-spine-fracture-detection\golden_dataset\det_train_manifest_balanced_1to1.csv',
    
    'model_name': 'tf_efficientnetv2_s.in21k_ft_in1k', 
    'img_size': 512,    
    'batch_size': 16,   
    'epochs': 100,
    'patience': 10,       
    'learning_rate': 5e-5, 
    'weight_decay': 5e-2,  
    
    'device': 'cuda' if torch.cuda.is_available() else 'cpu',
    'folds_to_run': [1],
    'save_dir': './runs/rsna_clf_efficientnet2'
}

# ==========================================
# 3. Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ï (Albumentations)
# ==========================================
train_transforms = A.Compose([
    A.Resize(CONFIG['img_size'], CONFIG['img_size']),
    A.HorizontalFlip(p=0.5), 
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.1, rotate_limit=10, p=0.5), 
    A.RandomBrightnessContrast(brightness_limit=0.15, contrast_limit=0.15, p=0.5), # üåû Î∞ùÍ∏∞/ÎåÄÎπÑ ÎûúÎç§ Î≥ÄÌôî
    A.GaussNoise(var_limit=(10.0, 50.0), p=0.3),                                   # üì∫ ÏßÄÏßÅÍ±∞Î¶¨Îäî ÎÖ∏Ïù¥Ï¶à Ï∂îÍ∞Ä
    A.CoarseDropout(max_holes=8, max_height=32, max_width=32, fill_value=0, p=0.3), 
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)), 
    ToTensorV2()
])

val_transforms = A.Compose([
    A.Resize(CONFIG['img_size'], CONFIG['img_size']),
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2()
])

# ==========================================
# 4. Îç∞Ïù¥ÌÑ∞ÏÖã
# ==========================================
class RSNAClfDataset(Dataset):
    def __init__(self, df, base_dir, transforms=None):
        self.df = df
        self.base_dir = base_dir
        self.transforms = transforms
        self.clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        
    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        full_path = os.path.join(self.base_dir, row['file_path'])
        
        try:
            with np.load(full_path) as loaded:
                img = loaded['data'].astype(np.float32) 
        except:
            img = np.zeros((3, 512, 512), dtype=np.float32)

        img_uint8 = (img * 255).astype(np.uint8)
        channels = []
        for i in range(3):
            channels.append(self.clahe.apply(img_uint8[i]))
        
        img_hwc = np.stack(channels, axis=-1) 
        
        if self.transforms:
            augmented = self.transforms(image=img_hwc)
            img_tensor = augmented['image']
        else:
            img_tensor = torch.from_numpy(np.transpose(img_hwc, (2,0,1))).float() / 255.0

        label = torch.tensor([float(row['fracture'])], dtype=torch.float32)
        return img_tensor, label

# ==========================================
# 5. Î™®Îç∏ Íµ¨ÏÑ±
# ==========================================
def build_model(model_name, pretrained=True):
    print(f"üõ†Ô∏è Î™®Îç∏ ÎπåÎìú Ï§ë: {model_name}")

    model = timm.create_model(
        model_name, 
        pretrained=pretrained, 
        in_chans=3, 
        num_classes=1,
        drop_rate=0.3,       # ÎßàÏßÄÎßâ Ï∂úÎ†•Ï∏µÏùò 30%Î•º Î¨¥ÏûëÏúÑÎ°ú ÎÅî
        drop_path_rate=0.2   # Ï§ëÍ∞Ñ Ï∏µÏùò 20%Î•º Î¨¥ÏûëÏúÑÎ°ú Í±¥ÎÑàÎúÄ
    )
    
    return model

# ==========================================
# 6. ÌïôÏäµ Î∞è Í≤ÄÏ¶ù Ìï®Ïàò
# ==========================================
def train_one_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    
    pbar = tqdm(dataloader, desc="Training", leave=False)
    for images, labels in pbar:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)  
        loss = criterion(outputs, labels) 
        
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        pbar.set_postfix({'loss': f"{loss.item():.4f}"})
        
    return running_loss / len(dataloader)

@torch.no_grad()
def validate(model, dataloader, criterion, device):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_targets = []
    
    pbar = tqdm(dataloader, desc="Validating", leave=False)
    for images, labels in pbar:
        images, labels = images.to(device), labels.to(device)
        
        outputs = model(images)
        loss = criterion(outputs, labels)
        running_loss += loss.item()
        
        probs = torch.sigmoid(outputs).cpu().numpy()
        targets = labels.cpu().numpy()
        
        all_preds.extend(probs)
        all_targets.extend(targets)
        
    all_preds = np.array(all_preds).flatten()
    all_targets = np.array(all_targets).flatten()
    
    preds_binary = (all_preds >= 0.5).astype(int)
    
    val_loss = running_loss / len(dataloader)
    
    # [ÌïµÏã¨] ÌèâÍ∞Ä ÏßÄÌëú 5Ï¢Ö ÏÑ∏Ìä∏ Í≥ÑÏÇ∞
    try:
        auc = roc_auc_score(all_targets, all_preds)
    except:
        auc = 0.5 
    acc = accuracy_score(all_targets, preds_binary)
    recall = recall_score(all_targets, preds_binary, zero_division=0)
    precision = precision_score(all_targets, preds_binary, zero_division=0) # Ï†ïÎ∞ÄÎèÑ Ï∂îÍ∞Ä
    f1 = f1_score(all_targets, preds_binary, zero_division=0)               # F1 Ï†êÏàò Ï∂îÍ∞Ä
    
    return val_loss, auc, acc, recall, precision, f1

# ==========================================
# 7. Î©îÏù∏ Ïã§Ìñâ
# ==========================================
if __name__ == '__main__':
    os.makedirs(CONFIG['save_dir'], exist_ok=True)
    
    df = pd.read_csv(CONFIG['manifest_path'])
    device = CONFIG['device']
    
    print(f"üöÄ [Classification Mode] ÌõàÎ†® ÏãúÏûë! (Model: {CONFIG['model_name']})")
    
    for fold in CONFIG['folds_to_run']:
        print(f"\n{'='*40}\nüåü FOLD {fold} ÏãúÏûë\n{'='*40}")
        
        train_df = df[df['fold'] != fold].reset_index(drop=True)
        val_df = df[df['fold'] == fold].reset_index(drop=True)
        
        train_ds = RSNAClfDataset(train_df, CONFIG['base_dir'], transforms=train_transforms)
        val_ds = RSNAClfDataset(val_df, CONFIG['base_dir'], transforms=val_transforms)
        
        train_loader = DataLoader(train_ds, batch_size=CONFIG['batch_size'], shuffle=True, num_workers=0, pin_memory=True)
        val_loader = DataLoader(val_ds, batch_size=CONFIG['batch_size'], shuffle=False, num_workers=0, pin_memory=True)
        
        model = build_model(CONFIG['model_name']).to(device)
        criterion = nn.BCEWithLogitsLoss()
        optimizer = torch.optim.AdamW(model.parameters(), lr=CONFIG['learning_rate'], weight_decay=CONFIG['weight_decay'])
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=CONFIG['epochs'])
        
        best_auc = 0.0
        early_stop_counter = 0 # ‚è±Ô∏è [Ï∂îÍ∞Ä] Ï∞∏ÏùÑÏÑ± Ïπ¥Ïö¥ÌÑ∞ ÏãúÏûë!
        
        # üìù [Ï∂îÍ∞Ä] 6ÎåÄ ÌèâÍ∞ÄÏßÄÌëúÎ•º Î™®Îëê Îã¥ÏùÑ Îπà Î∞îÍµ¨Îãà Ï§ÄÎπÑ
        history = {
            'epoch': [], 'train_loss': [], 'val_loss': [], 
            'val_auc': [], 'val_acc': [], 'val_recall': [],
            'val_precision': [], 'val_f1': []
        }
        
        for epoch in range(1, CONFIG['epochs'] + 1):
            print(f"\n[Epoch {epoch}/{CONFIG['epochs']}]")
            
            train_loss = train_one_epoch(model, train_loader, criterion, optimizer, device)
            # validate Ìï®ÏàòÏóêÏÑú ÏÉàÎ°ú Ï∂îÍ∞ÄÎêú precisionÍ≥º f1ÎèÑ Î∞õÏïÑÏòµÎãàÎã§.
            val_loss, val_auc, val_acc, val_recall, val_precision, val_f1 = validate(model, val_loader, criterion, device)
            
            scheduler.step() 
            
            print(f"üìâ Train Loss: {train_loss:.4f} | üìà Val Loss: {val_loss:.4f}")
            print(f"üèÜ AUC: {val_auc:.4f} | Acc: {val_acc:.4f} | üî• Recall: {val_recall:.4f} | üéØ Prec: {val_precision:.4f} | ‚öñÔ∏è F1: {val_f1:.4f}")
            
            # üìù Í∏∞Î°ùÏû•Ïóê Ï∞®Í≥°Ï∞®Í≥° Îã¥Í∏∞
            history['epoch'].append(epoch)
            history['train_loss'].append(train_loss)
            history['val_loss'].append(val_loss)
            history['val_auc'].append(val_auc)
            history['val_acc'].append(val_acc)
            history['val_recall'].append(val_recall)
            history['val_precision'].append(val_precision)
            history['val_f1'].append(val_f1)
            
            # --- üõë [ÏàòÏ†ï] ÏñºÎ¶¨ Ïä§ÌÉÄÌïë & Î™®Îç∏ Ï†ÄÏû• Î°úÏßÅ ---
            if val_auc > best_auc:
                best_auc = val_auc
                early_stop_counter = 0 # Ïã†Í∏∞Î°ù Îã¨ÏÑ±! Ïπ¥Ïö¥ÌÑ∞ Ï¥àÍ∏∞Ìôî
                save_path = os.path.join(CONFIG['save_dir'], f"best_efficientnet_fold{fold}.pth")
                torch.save(model.state_dict(), save_path)
                print(f"üíæ Best Model Ï†ÄÏû•Îê®! (AUC: {best_auc:.4f}) - Ïπ¥Ïö¥ÌÑ∞ Ï¥àÍ∏∞Ìôî üîÑ")
            else:
                early_stop_counter += 1 # Í∏∞Î°ù Í∞±Ïã† Ïã§Ìå®... Ïπ¥Ïö¥ÌÑ∞ Ï¶ùÍ∞Ä
                print(f"‚ö†Ô∏è Í∏∞Î°ù Í∞±Ïã† Ïã§Ìå®... (Early Stopping Ïπ¥Ïö¥ÌÑ∞: {early_stop_counter}/{CONFIG['patience']})")
                
                # Ïù∏ÎÇ¥Ïã¨Ïùò ÌïúÍ≥Ñ(patience)Ïóê ÎèÑÎã¨ÌïòÎ©¥ Í∞ïÏ†ú Ï¢ÖÎ£å!
                if early_stop_counter >= CONFIG['patience']:
                    print(f"\nüö® {CONFIG['patience']} ÏóêÌè≠ ÎèôÏïà ÏÑ±Îä• Ìñ•ÏÉÅÏù¥ ÏóÜÏñ¥ ÌïôÏäµÏùÑ Ï°∞Í∏∞ Ï¢ÖÎ£å(Early Stopping) Ìï©ÎãàÎã§!")
                    break # forÎ¨∏ ÌÉàÏ∂ú!
                    
        print(f"\n‚ú® FOLD {fold} ÌõàÎ†® ÏôÑÎ£å! ÏµúÍ≥† AUC: {best_auc:.4f}")
        
        # üìä ÏóëÏÖÄ(CSV) Ï†ÄÏû•
        history_df = pd.DataFrame(history)
        csv_path = os.path.join(CONFIG['save_dir'], f"history_fold{fold}.csv")
        history_df.to_csv(csv_path, index=False)
        print(f"üìä ÌïôÏäµ Í∏∞Î°ù CSV Ï†ÄÏû• ÏôÑÎ£å: {csv_path}")

        # üé® ÌõàÎ†® Í≥°ÏÑ† ÏãúÍ∞ÅÌôî Î∞è PNG Ï†ÄÏû• (ÏûòÎ¶∞ Î∂ÄÎ∂Ñ ÏàòÏ†ï ÏôÑÎ£å)
        plt.figure(figsize=(15, 5)) # Í∑∏ÎûòÌîÑ 3Í∞úÎ•º ÏúÑÌï¥ Í∞ÄÎ°úÎ•º Îçî ÎÑìÌòîÏäµÎãàÎã§.
        
        # 1. Loss Í∑∏ÎûòÌîÑ
        plt.subplot(1, 3, 1)
        plt.plot(history['epoch'], history['train_loss'], label='Train Loss', marker='o')
        plt.plot(history['epoch'], history['val_loss'], label='Val Loss', marker='o')
        plt.title(f'Fold {fold} - Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)

        # 2. Metric(AUC, Acc) Í∑∏ÎûòÌîÑ
        plt.subplot(1, 3, 2)
        plt.plot(history['epoch'], history['val_auc'], label='Val AUC', marker='o', color='red')
        plt.plot(history['epoch'], history['val_acc'], label='Val Acc', marker='o', color='purple')
        plt.title(f'Fold {fold} - AUC & Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Score')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)

        # 3. Metric(Recall, Precision, F1) Í∑∏ÎûòÌîÑ
        plt.subplot(1, 3, 3)
        plt.plot(history['epoch'], history['val_recall'], label='Recall', marker='o', color='orange')
        plt.plot(history['epoch'], history['val_precision'], label='Precision', marker='o', color='green')
        plt.plot(history['epoch'], history['val_f1'], label='F1-Score', marker='o', color='blue')
        plt.title(f'Fold {fold} - Recall / Prec / F1')
        plt.xlabel('Epoch')
        plt.ylabel('Score')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)

        plot_path = os.path.join(CONFIG['save_dir'], f"learning_curve_fold{fold}.png")
        plt.tight_layout()
        plt.savefig(plot_path, dpi=300)
        print(f"üìà Í∑∏ÎûòÌîÑ Ïù¥ÎØ∏ÏßÄ Ï†ÄÏû• ÏôÑÎ£å: {plot_path}")
        plt.show()
```

```python
import os
import torch
import pandas as pd
import numpy as np
import cv2
import timm
from tqdm import tqdm
from sklearn.metrics import roc_auc_score, accuracy_score, recall_score, precision_score, f1_score
import albumentations as A
from albumentations.pytorch import ToTensorV2
from torch.utils.data import Dataset, DataLoader
import os
import cv2
# --- Ïù¥ Îëê Ï§ÑÏùÑ importÎì§ Î∞îÎ°ú Î∞ëÏóê Íº≠ Ï∂îÍ∞ÄÌï¥Ï£ºÏÑ∏Ïöî! ---
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
cv2.setNumThreads(0)

# --- [1. ÏÑ§Ï†ï] ---
CONFIG = {
    'base_dir': r'C:\conda\3. Project\rsna-2022-cervical-spine-fracture-detection\golden_dataset',
    # üéØ [ÌïµÏã¨] Ïù¥Ï†ú 1:1 Îç∞Ïù¥ÌÑ∞Í∞Ä ÏïÑÎãàÎùº Ï†ÑÏ≤¥(Full) Îç∞Ïù¥ÌÑ∞Î•º Î∂àÎü¨ÏòµÎãàÎã§!
    'manifest_path': r'C:\conda\3. Project\rsna-2022-cervical-spine-fracture-detection\golden_dataset\det_train_manifest_cropped_full.csv',
    
    'model_name': 'tf_efficientnetv2_s.in21k_ft_in1k', 
    'weights_path': './runs/rsna_clf_efficientnet/best_efficientnet_fold1.pth', # Î∞©Í∏à Ï†ÄÏû•Îêú Î™®Îç∏
    'img_size': 512,
    'batch_size': 16, # Ï∂îÎ°†(ÌèâÍ∞Ä)Îßå ÌïòÎãàÍπå Î∞∞Ïπò ÏÇ¨Ïù¥Ï¶àÎ•º ÎÑâÎÑâÌïòÍ≤å!
    'device': 'cuda' if torch.cuda.is_available() else 'cpu'
}

# --- [2. ÌèâÍ∞ÄÏö© Ï†ÑÏ≤òÎ¶¨ Î∞è Îç∞Ïù¥ÌÑ∞ÏÖã] ---
val_transforms = A.Compose([
    A.Resize(CONFIG['img_size'], CONFIG['img_size']),
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2()
])

class RSNATestDataset(Dataset):
    def __init__(self, df, base_dir, transforms=None):
        self.df = df
        self.base_dir = base_dir
        self.transforms = transforms
        self.clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        
    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        full_path = os.path.join(self.base_dir, row['file_path'])
        
        try:
            with np.load(full_path) as loaded:
                img = loaded['data'].astype(np.float32) 
        except:
            img = np.zeros((3, 512, 512), dtype=np.float32)

        img_uint8 = (img * 255).astype(np.uint8)
        channels = [self.clahe.apply(img_uint8[i]) for i in range(3)]
        img_hwc = np.stack(channels, axis=-1) 
        
        if self.transforms:
            img_tensor = self.transforms(image=img_hwc)['image']
        else:
            img_tensor = torch.from_numpy(np.transpose(img_hwc, (2,0,1))).float() / 255.0

        label = float(row['fracture'])
        return img_tensor, label

# --- [3. Ï∂îÎ°† Î∞è Threshold ÌäúÎãù Ïã§Ìñâ] ---
def run_threshold_tuning():
    print("üõ†Ô∏è Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞(Full) Í≤ÄÏ¶ùÏÖã Ï§ÄÎπÑ Ï§ë...")
    
    # 1. Ï†ÑÏ≤¥ Îß§ÎãàÌéòÏä§Ìä∏ÏóêÏÑú Fold 1(Í≤ÄÏ¶ùÏÖã)Îßå Î∂ÑÎ¶¨
    df = pd.read_csv(CONFIG['manifest_path'])
    val_df = df[df['fold'] == 1].reset_index(drop=True)
    
    val_ds = RSNATestDataset(val_df, CONFIG['base_dir'], transforms=val_transforms)
    val_loader = DataLoader(val_ds, batch_size=CONFIG['batch_size'], shuffle=False, num_workers=0)
    
    # 2. Î™®Îç∏ Î°úÎìú
    model = timm.create_model(CONFIG['model_name'], pretrained=False, in_chans=3, num_classes=1)
    model.load_state_dict(torch.load(CONFIG['weights_path'], map_location=CONFIG['device']))
    model.to(CONFIG['device'])
    model.eval()
    
    # 3. Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÌôïÎ•†Í∞í(0~1) ÎΩëÏïÑÎÇ¥Í∏∞
    all_preds = []
    all_targets = []
    
    print("üöÄ Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ Ï∂îÎ°† ÏãúÏûë! (Ïû†ÏãúÎßå Í∏∞Îã§Î†§Ï£ºÏÑ∏Ïöî...)")
    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc="Inference"):
            images = images.to(CONFIG['device'])
            outputs = model(images)
            probs = torch.sigmoid(outputs).cpu().numpy().flatten()
            
            all_preds.extend(probs)
            all_targets.extend(labels.numpy())
            
    all_preds = np.array(all_preds)
    all_targets = np.array(all_targets)
    
    # Í∏∞Î≥∏ AUC Í≥ÑÏÇ∞ (ThresholdÏôÄ Î¨¥Í¥ÄÌïú Í≥†Ïú† Ïã§Î†•)
    full_auc = roc_auc_score(all_targets, all_preds)
    print(f"\nüåü [ÏµúÏ¢Ö ÌèâÍ∞Ä] Full Data Validation AUC: {full_auc:.4f}\n")
    
    # 4. Threshold 0.1 Î∂ÄÌÑ∞ 0.9 ÍπåÏßÄ ÎèåÎ†§Î≥¥Î©∞ Í≤∞Í≥º ÎπÑÍµê
    print("="*60)
    print(f"{'Threshold':^10} | {'Recall':^8} | {'Precision':^10} | {'F1-Score':^8} | {'Accuracy':^8}")
    print("-"*60)
    
    for th in np.arange(0.1, 1.0, 0.1):
        preds_binary = (all_preds >= th).astype(int)
        
        rec = recall_score(all_targets, preds_binary, zero_division=0)
        pre = precision_score(all_targets, preds_binary, zero_division=0)
        f1 = f1_score(all_targets, preds_binary, zero_division=0)
        acc = accuracy_score(all_targets, preds_binary)
        
        # Î≥¥Í∏∞ Ï¢ãÍ≤å ÌïòÏù¥ÎùºÏù¥Ìä∏ (RecallÏù¥ 0.8 Ïù¥ÏÉÅÏù¥Î©¥ÏÑú F1Ïù¥ Ï†úÏùº Î¨¥ÎÇúÌïú Í≥≥ Ï∞æÍ∏∞)
        print(f"   {th:.1f}      |  {rec:.4f}  |   {pre:.4f}   |  {f1:.4f}  |  {acc:.4f}")
    print("="*60)
    
    # ÏòàÏ∏°Í∞í CSV Ï†ÄÏû• (Îã§Ïùå Ïä§ÌÖùÏù∏ Seg ÎßàÏä§ÌÅ¨ Í≤∞Ìï©ÏùÑ ÏúÑÌï¥ Íº≠ ÌïÑÏöî!)
    val_df['pred_prob'] = all_preds
    save_csv = './runs/rsna_clf_efficientnet/val_full_predictions.csv'
    val_df.to_csv(save_csv, index=False)
    print(f"\nüíæ ÏòàÏ∏° Í≤∞Í≥º(ÌôïÎ•†Í∞í)Í∞Ä Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§: {save_csv}")
    print("üëâ (Ïù¥ ÌååÏùºÏù¥ Îã§Ïùå Îã®Í≥ÑÏù∏ Seg ÎßàÏä§ÌÅ¨ÏôÄÏùò Í≤∞Ìï©Ïóê ÏÇ¨Ïö©Îê©ÎãàÎã§!)")

if __name__ == '__main__':
    run_threshold_tuning()
```
